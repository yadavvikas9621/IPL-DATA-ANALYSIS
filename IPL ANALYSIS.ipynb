```python
#Which Team had won by (closest margin) minimum runs?
#To know the team that was won by the closest margin, 
#we have to do the opposite of what weâ€™ve done in the above steps. But thereâ€™s a caveat in that. 
##In Cricket, if the team thatâ€™s batting second wins the match (by chasing), 
#itâ€™ll be said â€œTeam second won by x number of wicketsâ€ and if the team thatâ€™s bowling second (batted first) wins the match (by defending their score), itâ€™ll be said â€œTeam first won by x number of runsâ€. Hence, the minimum win by runs will always be 0 and the minimum win by wickets will also always be 0 in a tournament since sometimes chasing team or sometimes the team that batted first could win. To overcome this caveat, we just have to apply 
#a simple workaround as you can see below.
r.iloc[r[r['win_by_runs'].ge(1)].win_by_runs.idxmin()]['winner']
```

```python
##Which Team had won by minimum wickets?
r.iloc[r[r['win_by_wickets'].ge(1)].win_by_wickets.idxmin()]
```
